```python
import logging

logging.getLogger().setLevel(logging.ERROR)

import apache_beam as beam
from apache_beam.runners.interactive import interactive_runner
from apache_beam.runners.interactive.display import pipeline_graph

def display_pipeline(pipeline):
    pipeline_graph.PipelineGraph(pipeline).display_graph()

p = beam.Pipeline(interactive_runner.InteractiveRunner())
init = p |  'Init' >> beam.Create(range(10))
```

```python
square = init | beam.Map(lambda x: x*x)
display_pipeline(p)
```

```python
cube = init | beam.Map(lambda x: x**3)
```

```python
fourth_power = square | 'Square' >> beam.Map(lambda x: str(x))
fourth_power = init | beam.Map(lambda x: x*x) | beam.Map(lambda x: x*x)
```

```python
square = init | beam.Map(lambda x: x*x)
fourth_power = square | beam.Map(lambda x: x*x)
```

```python
result = p.run()
result.wait_until_finish()
```

```python
print(result.get(square))
print(result.get(fourth_power))
```
